#1) Convert string column into date datatype

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv",parse_dates=["day"])
type(df.day[0])
```

#3) Make day as the index column

```python
df.set_index('day',inplace=True)
df
```

#4)fillna() method
###fills nil with a certain value

```python
new_df=df.fillna(0)
new_df
```

```python
new_df1=df.fillna({
    'tempereture':0,
    'windspeed':0,
    "event":"no event"

})
new_df1
```

```python
df.temperature.mean()
```

#5) fillna(method="ffill")
###fills na with top row entry value

```python
new_df2=df.fillna(method="ffill")
new_df2
```

```python
new_df6=df.fillna(method="ffill",limit=1)
new_df6

```

#6)fillna(method="bfill")
###fills na with next below row entry value

```python
new_df4=df.fillna(method="bfill")
new_df4
```

### fills na with nest column entry value

```python
new_df5=df.fillna(method="bfill",axis="columns")
new_df5
```

#7)interpolate()

```python
new_df7=df.interpolate()
new_df7
```

```python
new_df8=df.interpolate(method="time")
new_df8
```

#8) Delete rows with any na value

```python
new_df9=df.dropna()
new_df9
```

###drops rows that has all values na

```python
new_df10=df.dropna(how="all")
new_df10
```

###atleast n na values, keep the row

```python
new_df10=df.dropna(thresh=3)
new_df10
```

#9) Fill in the missing dates

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
new_df11 = df.reindex(idx)
new_df11
```
