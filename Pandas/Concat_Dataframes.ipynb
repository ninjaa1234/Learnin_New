```python
import pandas as pd
```

```python
india_weather=pd.DataFrame({
    "city":["mumbai",'delhi','banglore'],
    'temperature':[32,45,60],
    'humidity':[80,60,70]
})
india_weather
```

```python
usa_weather=pd.DataFrame({
    "city":["new_york",'chicago','orlando'],
    'temperature':[23,14,35],
    'humidity':[80,60,70]
})
usa_weather
```

#Concat with continous index

```python
pd.concat([india_weather,usa_weather],ignore_index=True)
```

#Concat with keys

```python
df=pd.concat([india_weather,usa_weather],keys=["india","usa"])
```

```python
df.loc['india']
```

#Concat two datasets with each other as columns

```python
city_temp=pd.DataFrame({
    "city":["mumbai",'delhi','banglore'],
    'temperature':[32,45,60]
})
city_temp
```

```python
windspeed=pd.DataFrame({
    'humidity':[80,60,70]
})
windspeed
```

```python
df=pd.concat([city_temp,windspeed],axis="columns")
df
```
