```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
```

```python
dir(digits)
```

```python
digits=load_digits()
```

```python
#display the corresponding data in figures
plt.gray()
plt.matshow(digits.images[0])
```

```python
df=pd.DataFrame(digits.data)        
df.head()
```

```python
df["target"]=digits.target
df
```

```python
df["target_names"]=df.target.apply(lambda x:digits.target_names[x])
df.head()
```

```python
X=df.drop(['target','target_names'],axis="columns")     #independent variables
X.head()
```

```python
Y=pd.DataFrame(digits.target)           #dependent variables
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.svm import SVC
reg=SVC(kernel="linear")
reg.fit(X_train,Y_train)
reg.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
reg=SVC(kernel="rbf")
reg.fit(X_train,Y_train)
reg.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
reg=SVC(gamma=1000)
reg.fit(X_train,Y_train)
reg.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
reg=SVC(C=2)
reg.fit(X_train,Y_train)
reg.score(X_test,Y_test)
```
