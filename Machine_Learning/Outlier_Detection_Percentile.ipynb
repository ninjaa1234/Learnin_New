```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/1_outliers/heights.csv")
df.head()
```

#Example 1

#percentile thresold

```python
max_thresold=df.height.quantile(0.95)            #all hieghts above 9.689 are 95 percentile
max_thresold
```

```python
df[df['height']>max_thresold]          #no man can have a heigth of 14 ft
```

```python
min_thresold=df.height.quantile(0.05)
min_thresold
```

```python
df[df['height']<min_thresold]
```

```python
df[(df['height']<max_thresold) & (df['height']>min_thresold)]
```

#Example 2

```python
df1=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/1_outliers/bhp.csv")
df1.head()
```

```python
df1.shape
```

```python
df1.describe()              #75cent of houses has sq feet less than 1672 value
```

```python
min_thresold1, max_thresold1 = df1.price_per_sqft.quantile([0.001, 0.999])
min_thresold1, max_thresold1
```

```python
df1[df1.price_per_sqft < min_thresold1]         #you cant get a home of 371 in Bangalore
                                                #These are more likely errors
```

```python
df1[df1.price_per_sqft > max_thresold1]
```

```python
df2=df1[(df1.price_per_sqft < max_thresold1) & (df1.price_per_sqft > min_thresold1)]
df2.shape
```

```python
df2.sample(10)#to print 10 random rows
```
