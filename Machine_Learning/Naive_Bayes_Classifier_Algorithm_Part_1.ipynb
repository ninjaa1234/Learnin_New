```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/titanic.csv")
df.head()
```

```python
df=df.drop(['Name','PassengerId','SibSp','Parch','Ticket','Cabin','Embarked'],axis="columns")
df.head()           #no impact on the survival rate
```

```python
X=df.drop(['Survived'],axis='columns')
Y=df.Survived
```

```python
dummies=pd.get_dummies(X.Sex)
dummies.head()
```

```python
merged=pd.concat([X,dummies],axis="columns")
merged.head()
```

```python
final=merged.drop(['Sex','female'],axis="columns")
final.head()
```

```python
X=final
```

```python
X.columns[X.isna().any()]          #to check for any na values
```

```python
X.Age=X.Age.fillna(X.Age.mean())        #fillna with mean values in age column
X.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

#Gaussian Naive Baise Model

```python
from sklearn.naive_bayes import GaussianNB
reg=GaussianNB()
reg.fit(X_train,Y_train)
reg.score(X_test,Y_test)
```
