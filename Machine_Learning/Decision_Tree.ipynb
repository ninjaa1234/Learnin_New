```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
df.head()
```

#Seperate Dependent and independent variables

```python
#independent variables
X=df.drop('salary_more_then_100k',axis='columns')
#dependent variables
Y=df['salary_more_then_100k']
```

```python
X.head()
```

```python
Y.head()
```

#Remove text information

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
X['company_n'] = le_company.fit_transform(X['company'])
X['job_n'] = le_job.fit_transform(X['job'])
X['degree_n'] = le_degree.fit_transform(X['degree'])
X
```

```python
X1=X.drop(['company','job','degree'],axis='columns')
X1
```

#Descision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
reg=DecisionTreeClassifier()
```

```python
reg.fit(X1,Y)
```

```python
reg.score(X1,Y)
```
