```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv")
df
```

```python
dummies=pd.get_dummies(df["Car Model"])
dummies
```

```python
merged=pd.concat([df,dummies],axis="columns")
merged
```

```python
final=merged.drop(['Car Model','Audi A5'],axis='columns')
final
```

```python
reg=linear_model.LinearRegression()
```

```python
X=final.drop(['Sell Price($)'],axis='columns')
X
```

```python
Y=final['Sell Price($)']
Y
```

```python
reg.fit(X,Y)
```

```python
#The price of Benz 4 years old mileage 45000
reg.predict([[45000,4,0,1]])
```

```python
#The price of BMW 7 years old mileage 8600
reg.predict([[86000,7,1,0]])
```

```python
#the accuracy of prediction
reg.score(X,Y)
```
