```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv")
df.head()
```

#Find the survival rate of Passengers(neglect all the red columns)

```python
#dependent variable
Y=df.Survived
```

```python
del_col=df.drop(["Survived","PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis="columns")
#independent columns
X=del_col
X.head()
```

```python
dummies=pd.get_dummies(X.Sex)
dummies
```

```python
merged=pd.concat([X,dummies],axis="columns")
merged
```

```python
final=merged.drop(["Sex","female"],axis="columns")
final
```

```python
final=final.fillna(0)
final
```

```python
X=final
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
reg=DecisionTreeClassifier()
```

```python
X
```

```python
Y
```

```python
reg.fit(X,Y)
```

```python
reg.score(X,Y)
```

```python
df=pd.concat([X,Y],axis="columns")
df
```
