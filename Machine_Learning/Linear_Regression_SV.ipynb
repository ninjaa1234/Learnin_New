```python
#import all the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv')
df
```

```python
#plot the graph

plt.plot(df.area,df.price,'r+',markersize=10)
```

```python
plt.xlabel('area(sq.ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
#create a object for linear regression
reg=linear_model.LinearRegression()
#fit the data
reg.fit(df[['area']],df.price)
```

#1)Predict the price of house less than 3300 sq feet

```python
#predict the prices of the houses under 3300square feet
reg.predict(([[3300]]))
```

#Checking using mathematical method
y=m*x+b

```python
#x is the area to predict
x=3300
```

```python
#m is the slope aka coefficient
m=reg.coef_
m
```

```python
##b is the y intercept
b=reg.intercept_
b
```

```python
y=m*x+b
y
```

#Plot the predicted price

```python
plt.xlabel('area(sq.ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color="blue")
```
