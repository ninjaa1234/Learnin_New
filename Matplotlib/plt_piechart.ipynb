```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

#increase the radius of the chart

```python
plt.pie(exp_vals,labels=exp_labels,radius=2)
plt.show()
```

#autopct argument for percentage

```python
#no of decimal poins of 2 is %0.2f%%
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,0.5,0,0,0])
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,0.5,0,0,0],startangle=45)
plt.show()
```
